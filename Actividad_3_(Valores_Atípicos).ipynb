```python
data=pd.read_csv("limpio.csv")
data.head(5)
```

```python
#Eliminamos la columnas Unnamed: 0
data2=data.drop(["Unnamed: 0"],axis=1)
data2
```

```python
#Verificamos que no existan nulos
print("Datos nulos: ",data2.isnull().sum().sum())
print(data2.info())
```

```python
#Seleccionamos solo las columnas de tipo float para hacer nuestro boxplot por df
diagrama=data2.iloc[:,1:24]
diagrama
```

```python
#Diagrama de caja y bigote
fig=plt.figure(figsize=(10,6))
plt.boxplot(diagrama,vert=False)
plt.title("Outliers del Dataframe")
plt.show()
```

#ventas_precios_corrientes

```python
#ventas_precios_corrientes
fig=plt.figure(figsize=(5,1))
plt.boxplot(data2["ventas_precios_corrientes"],vert=False)
plt.title("Outliers de ventas_precios_corrientes")
plt.show()
```

```python
#Metodo de IQR
y=data2["ventas_precios_corrientes"]
percentile25=y.quantile(0.25)#Q1
percentile75=y.quantile(0.75)#Q3

iqr=percentile75-percentile25

limite_sup=percentile75+1.5*iqr
limite_infe=percentile25-1.5*iqr

print("Limite superior: ",limite_sup)
print("Limite inferior: ",limite_infe)
```

```python
#Obtenemos un dataframe sin outliers
data_clean=data2[(y<=limite_sup)&(y>=limite_infe)]
print(data_clean.shape)

#Comprobamos
fig=plt.figure(figsize=(5,4))
plt.boxplot(data_clean["ventas_precios_corrientes"],vert=False)
plt.title("ventas_precios_corrientes limpio")
plt.show()
```
